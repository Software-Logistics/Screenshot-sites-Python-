```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import WebDriverException
import time
import random
import sys
import pandas as pd
import pyautogui

#Input your Excel sheet------------------------------------
df=pd.read_csv(r'ADD HERE THE CSV FILE FOR THE SITE')
#Read the Excel Sheet -----------------------------------
a=df["Now add here company's url column"]
options = webdriver.ChromeOptions() 
options.add_argument(r"--user-data-dir=C:\Users\Admin\AppData\Local\Google\Chrome\User Data\Default") #Path to your chrome profile
#options.add_argument(r'--profile-directory=C:\Users\Admin\AppData\Local\Google\Chrome\User Data\Profile 1')
#INPUT PATH FOR YOUR CHROME DRIVER--------------------------------------------------------------------------
driver= webdriver.Chrome(executable_path=r"C:\Users\Admin\Desktop\codes\chromedriver.exe", options=options)
driver.maximize_window()
for x in range(1,len(a)):
    try:
        
        driver.get('https://www.'+a[x])
        time.sleep(5)
        ActionChains(driver).send_keys(Keys.ESCAPE).perform()
        time.sleep(4)
        myScreenshot = pyautogui.screenshot()
        path_to_save=r"C:\Users\Admin\Desktop\codes\New folder\1"+str(a[x])+".png"
        myScreenshot.save(path_to_save)
        tags=['clothing','Clothing','CLOTHING', 'Dress','clothes','Dresses','dress','Dress','tools','Tools','Products','Product','Gear ','Gear','Dresses','Shop']
        any_xpath=[]
        def search():
            
            for y in tags:
                try:
                    driver.find_element_by_xpath("//*[contains(text(), '"+y+"')]").click()
                    any_xpath.append(y)
                    return 'ok1'
                except Exception:

                    try:
                        elements=driver.find_elements_by_xpath("//*[contains(text(), '"+y+"')]")
                        for element in elements:
                            try:
                                element.click()
                                return 'ok2'
                            except Exception:
                                continue
                    except Exception:
                        continue
        search()           
        time.sleep(15)
        print(any_xpath)
        #tags_for_product=['dress',"Dress",""]
    except WebDriverException:
        continue
        
```
